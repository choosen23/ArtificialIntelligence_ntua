Συνεπώς η συνάρτηση που θα κάνει τις προτάσεις (recommendation) με είσοδο μια ταινία πρέπει να επιστρέφει-εκτυπώνει μια λίστα με τις προτεινόμενες ταινίες σε φθίνουσα σειρά ομοιότητας. 

Για να μπορείτε να αξιολογήσετε τα αποτελέσματα των συστάσεων σας σας δίνεται ένα αρχείο με ορισμένες ταινίας και τα 10 πρώτα αποτελέσματά τους σε φθίνουσα σειρά ομοιότητας. $$jaccard(A,B)=\frac{|A\cap B|}{|A\cup B|}$$

Η υλοποίηση της συνάρτησης αυτής σε Python σας δίνεται με το όνομα jaccard_similarity (set1, set2). 

Αξίζει να σημειωθεί ότι η παραπάνω σύγκριση δεν είναι απόλυτη για τον λόγο ότι οι συναρτήσεις ομοιότητας μπορεί να είναι διαφορετικές (διαφορετικά ορίζει ο καθένας της ομοιότητα μεταξύ ταινιών) καθώς επίσης αν μια ταινία δεν έχει πολύ σχετικές προτάσεις τότε οι πιο γενικές συναρτήσεις ομοιότητας θα έχουν πολλά αποτελέσματα τα όποια η Prolog θα επιστρέψει με την σειρά που τα βρίσκει στην βάση, όποτε τότε δεν έχει ιδιαίτερο νόημα η σύγκριση σε αυτήν την περίπτωση.
Η ομοιότητα των ταινιών είναι βοηθητική και σκοπό έχει να σας βοηθήσει να κατανοήσετε την κατεύθυνση της ασκήσης και όχι για να καταφέρετε την πλήρη ταύτιση των συστάσεων.

### 3ο Μέρος: Recommendation System με βάση τις προτιμήσεις-αξιολογήσεις του χρήστη


Σε αυτό το σημείο θα εργαστείτε με τα αρχεία user1.csv και user2.csv, τα όποια περιέχουν αξιολογήσεις (από 1 μέχρι 5) για ταινίες από 2 διαφορετικούς χρήστες. Το προηγούμενο σύστημα συστάσεων προτείνει στον χρήστη ταινίες αποκλειστικά με βάση την ομοιότητά τους. Έτσι σε αυτό το σημείο θα γίνει μια αναβάθμιση του συστήματος έτσι ώστε να παράγονται καλύτερες  συστάσεις οι όποιες θα λαμβάνουν υπόψιν και τις προτιμήσεις του χρήστη, οι όποιες θα εξάγονται από τις αξιολογήσεις που έχει κάνει μέχρι στιγμής. 

Έτσι πλέον για κάθε ταινία θα υπάρχει ένα score το όποιο αρχικά θε είναι ίσο με 0 και θα διαμορφώνεται από τις αξιολογήσεις κάθε user. Έτσι για έναν χρήστη με βάση τις αξιολογήσεις που έχει κάνει θα πρέπει για κάθε ταινία που έχει βαθμολογήσει θα βρίσκονται οι κοινές ταινίες ανά κλίμακα και στο μέχρι τώρα σκορ κάθε παρόμοιας ταινίας θα προστίθεται ένα βάρος το όποιο θα μπορούσε να είναι το ποσοστό ομοιότητας της ταινίας (δηλαδή ένα βάρος για κάθε κλίμακα) επί τον βαθμό που έχει βάλει ο χρήστης για την αρχική ταινία (ή ένα βάρος ανά βαθμό ώστε οι καλές κριτικές να γίνονται καλύτερες ενώ οι κακές χειρότερες). Δηλαδή, αν $rated\_movies$ είναι το σύνολο ταινιών που έχει αξιολογήσει ο χρήστης, τότε τα recommendations που θα λαμβάνει ο χρήστης θα είναι οι ταινίες που είναι τουλάχιστον similar_1 με τουλάχιστον μία ταινία που έχει βαθμολογήσει ο χρήστης, σε φθίνουσα σειρά recommendation_score, όπου:

$$recommendation\_score(m_1)=\sum_{m_2\in rated\_movies} similarity\_weight(m_1,m_2)\cdot voting\_weight(m_2)$$

````python
from pyswip import Prolog
import pandas as pd 
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv("movie_metadata.csv") 
#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
data.head()
````

````python
def clean_text(text):
    text = text.replace(u'\xa0', u'')
    text = text.replace(u"'", u'')
    text = text.replace(u'"', u'')
    return text
````

````python
#create World
#Ορίζουμε τον κόσμο μας
prolog = Prolog()
#και σβήνουμε ότι υπήρχε αποθηκευμένο από πρίν
prolog.query('make')
    
````

````python

#Για κάθε row του πίνακα φτιάχνουμε τα κατηγοτήματα που θέλουμε να αποθηκέυσουμε
#αρχικά σε μια λίστα με το όνομα literals
literals = []
movie_score = {}
for row in data.itertuples(index=True, name='Pandas'):
    movie_title = clean_text(getattr(row, 'movie_title'))
    director = clean_text(getattr(row, 'director_name'))
    color = clean_text(getattr(row,'color'))
    imdb_score = getattr(row,'imdb_score')
    actor_1_name = clean_text(getattr(row,'actor_1_name'))
    actor_2_name = clean_text(getattr(row,'actor_2_name'))
    actor_3_name = clean_text(getattr(row,'actor_3_name'))
    language = clean_text(getattr(row,'language'))
    content_rating = clean_text(getattr(row,'content_rating'))
    
    

    movie_score[movie_title] = 0
    
    if movie_title != "UNK":
        #these which have one attribute
        if director != "UNK":
            literals.append("director('"+ movie_title +"','" + director + "')")
        if (color != "UKN"): 
            literals.append("color('"+ movie_title +"','" + color + "')")
        if (imdb_score != "UKN"):
            literals.append("imdb_score('"+ movie_title +"','" + str(imdb_score) + "')")
        if (actor_1_name != "UKN"):
            literals.append("actor_1_name('"+ movie_title +"','" + actor_1_name + "')")
        if (actor_2_name != "UKN"):
            literals.append("actor_2_name('"+ movie_title +"','" + actor_2_name + "')")
        if (actor_3_name != "UKN"):
            literals.append("actor_3_name('"+ movie_title +"','" + actor_3_name + "')")
        if (language != "UKN"):
            literals.append("language('"+ movie_title +"','" + language + "')")
        if (content_rating != "UKN"):
            literals.append("content_rating('"+ movie_title +"','" + content_rating + "')")
            
            
        #these which have many
            
        #we create a cause for each genre
        for genre in getattr(row, 'genres').split("|"):
            literals.append("genre('"+ movie_title +"','"+ genre +"')")
        #we create a cause for each keyword
        for plot_keywords in getattr(row,'plot_keywords').split("|"):
            literals.append("plot_keywords('"+ movie_title +"','"+plot_keywords.replace("'"," ") +"')")
            
            

#Η Prolog θέλει τα κατηγορήματά της με την σειρά 
literals.sort()

#so as we wrote the file we just  ignore them
f = open("queries.pl", "w+")


#we save our clauses to part1a.pl file so we don't have to run it all the time
for literal in literals:
    prolog.assertz(literal)
    f.write(literal+"\n")
f.close()

#Επίσης μπορούμε να κάνουμε consult ένα έτοιμό αρχείο στον κόσμο όπως γίνεται παρακάτω
#prolog.consult("part1b.pl")


````

````python
def jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    union = len(set1) + len(set2) - intersection
    return intersection / union
````

````python
def sophisticated_recommender(movie):
    #have 4 common genre words
    q = prolog.query("find_similar_movies_1('" + movie +"',M)")
    ans1 = []
    for soln in q:
            ans1.append(soln["M"])
    q.close()
    ans1 = list(set(ans1))

    
    #have 3 common genre words
    q = prolog.query("find_similar_movies_2('" + movie +"',M)")
    ans2 = []
    for soln in q:
            ans2.append(soln["M"])
    q.close()
    ans2 = list(set(ans2))
    
    #have 1 common genre words
    q = prolog.query("find_similar_movies_3('" + movie +"',M)")
    ans3 = []
    for soln in q:
            ans3.append(soln["M"])
    q.close()
    ans3 = list(set(ans3))

    
    #the result for genres:
    if ( not ans1 ):
        genre_res = ans1
    else:
        if (not ans2 ):
            genre_res = ans2
        else:
            genre_res = ans3
    
        
    #same director
    q = prolog.query("find_similar_movies_4('" + movie +"',M)")
    ans4 = []
    for soln in q:
        ans4.append(soln["M"])
    q.close()
    ans4 = list(set(ans4))
    
    #results from director
    dir_res = ans4
    
    
    #have 4 common plot words
    q = prolog.query("find_similar_movies_5('" + movie +"',M)")
    ans5 = []
    for soln in q:
            ans5.append(soln["M"])
    q.close()            
    ans5 = list(set(ans5))
    
    #have 2 common plot words
    q = prolog.query("find_similar_movies_6('" + movie +"',M)")
    ans6 = []
    for soln in q:
            ans6.append(soln["M"])
    q.close()
    ans6 = list(set(ans6))
    
    #plot keywords results
    if (not ans5):
        plot_res = ans5
    else:
        plot_res = ans6
    
    
    #have 3 main actors in common
    q = prolog.query("find_similar_movies_7('" + movie +"',M)")
    ans7 = []
    for soln in q:
            ans7.append(soln["M"])
    q.close()
    ans7 = list(set(ans7))
    
    #have 2 main actor common
    q = prolog.query("find_similar_movies_8('" + movie +"',M)")
    ans8 = []
    for soln in q:
            ans8.append(soln["M"])
    q.close()
    ans8 = list(set(ans8))
    
    #have 1 main actor common
    q = prolog.query("find_similar_movies_9('" + movie +"',M)")
    ans9 = []
    for soln in q:
            ans9.append(soln["M"])
    q.close()
    ans9 = list(set(ans9))
    
    #the result for actors:
    if ( not ans7 ):
        act_res = ans7
    else:
        if (not ans8 ):
            act_res = ans8
        else:
            act_res = ans9
    
    
    
    #same language and same genre
    q = prolog.query("find_similar_movies_10('" + movie +"',M)")
    ans10 = []
    for soln in q:
            ans10.append(soln["M"])
    q.close()
    ans10 = list(set(ans10))
    
    lang_res = ans10
    
    #same genre and same color
    q = prolog.query("find_similar_movies_11('" + movie +"',M)")
    ans11 = []
    for soln in q:
        ans11.append(soln["M"])
    q.close()
    ans11 = list(set(ans11))
    
    col_res = ans11
    
    #same content rating
    q = prolog.query("find_similar_movies_12('" + movie +"',M)")
    ans12 = []
    for soln in q:
            ans12.append(soln["M"])
    q.close()
    ans12 = list(set(ans12))
    
    paradv_res = ans12
        
    standars = set(paradv_res + col_res + lang_res)
    ansb = list(set(act_res).intersection(standars))
    ansc = list(set(plot_res + genre_res + ansb + dir_res))
    ans = list(set(ansc))
        
        
    #κάνουμε κάποιο Processing στα παραπάνω αποτελέσματα και επιστρέφουμε τις "όμοιες" ταινίες με φθίνουσα σειρά
    return list(set(ans))
````

````python
def compute_scores(file):
    similarity_weights = {
        "similarity_1": 13,
        "similarity_2": 9,
        "similarity_3": 2,
        "similarity_4": 12, 
        "similarity_5": 14,
        "similarity_6": 9,
        "similarity_7": 16,
        "similarity_8": 13,
        "similarity_9": 7,
        "similarity_10": 10,
        "similarity_11": 10,
        "similarity_12": 12,    
        "similarity_13": 8
    }

    rating_weight = {5: 10, 4: 5, 3: 0, 2: -5, 1: -10}
    #gia kathe tainia vazoume ena score 0
    movie_score = {m: 0 for m in  movie_score}

    for row in file.itertuples(index=True, name='Pandas'):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')    


        q = prolog.query("find_similar_movies_1('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_1'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_2('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_2'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_3('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_3'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_4('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_4'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_5('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_5'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_6('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_6'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_7('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_7'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_8('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_8'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_9('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_9'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_10('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_10'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_11('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_11'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_12('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_12'] * rating_weight[rating] 
        q.close()

        q = prolog.query("find_similar_movies_13('" + movie + "',M)")
        for soln in q:
            movie_score [soln["M"]] += similarity_weights['similarity_13'] * rating_weight[rating] 
        q.close()
````
